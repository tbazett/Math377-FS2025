```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import quad

K = 14   # 14 vs 20
tau = 1.5  # 1.5 vs 1.75

N = K * 40
A = np.zeros(N)
B = np.zeros(N)
C = np.zeros(N)
D = np.zeros(N)
E = np.zeros(N)
F = np.zeros(N)
G = np.zeros(N)
H = np.zeros(N)

y = np.linspace(0, 20, N)
fun = lambda x: 0.2 * x * np.exp((1 - x) / 1)

# Set a function for the first car. Note that quad standards for quadrature, i.e think of this as integrating fun. 
for t in range(N):
    val, _ = quad(fun, 0, t / K)
    A[t] = -100000 * val / 3600

# All other cards follow
for t in range(int(K * tau), N - 1):
    B[t+1] = B[t] + 100000 * np.log(1 + 14.715 * (A[t-int(K*tau)] - B[t-int(K*tau)]) / 1000) / (3600*K)

for t in range(int(K * tau), N - 1):
    C[t+1] = C[t] + 100000 * np.log(1 + 14.715 * (B[t-int(K*tau)] - C[t-int(K*tau)]) / 1000) / (3600*K)

for t in range(int(K * tau), N - 1):
    D[t+1] = D[t] + 100000 * np.log(1 + 14.715 * (C[t-int(K*tau)] - D[t-int(K*tau)]) / 1000) / (3600*K)

for t in range(int(K * tau), N - 1):
    E[t+1] = E[t] + 100000 * np.log(1 + 14.715 * (D[t-int(K*tau)] - E[t-int(K*tau)]) / 1000) / (3600*K)

for t in range(int(K * tau), N - 1):
    F[t+1] = F[t] + 100000 * np.log(1 + 14.715 * (E[t-int(K*tau)] - F[t-int(K*tau)]) / 1000) / (3600*K)

for t in range(int(K * tau), N - 1):
    G[t+1] = G[t] + 100000 * np.log(1 + 14.715 * (F[t-int(K*tau)] - G[t-int(K*tau)]) / 1000) / (3600*K)

for t in range(int(K * tau), N - 1):
    H[t+1] = H[t] + 100000 * np.log(1 + 14.715 * (G[t-int(K*tau)] - H[t-int(K*tau)]) / 1000) / (3600*K)


plt.figure(figsize=(10,6))
plt.plot(y, B - 19, color='b', linewidth=3)
plt.plot(y, A, color=(1,0.5,0), linewidth=3)
plt.plot(y, C - 38, color='y', linewidth=3)
plt.plot(y, D - 57, color='m', linewidth=3)
plt.plot(y, E - 76, color='g', linewidth=3)
plt.plot(y, F - 95, color='c', linewidth=3)
plt.plot(y, G - 114, color='r', linewidth=3)
plt.plot(y, H - 133, linewidth=3)

plt.axis([0, 20, -170, 5])

plt.show()
```
