The first segment imports the various libraries we will use. This doesn't need to be changed for the homework. 

```python
import matplotlib.pyplot as plt #Matplotlib is used to generate our plots
from scipy.integrate import solve_ivp #We will use solve_ivp to numerically solve the system of differential equations. 
```

Next up we define the system of ODEs. You can expand the list of variables to represent your new system of ODEs. 

```python
# Define the system of ODEs
def sirsys1(t, y): #y is a vector whose components are Susceptibles, Infected, Recovered respectively
    S, I, R = y
    dS_dt = -a * S * I
    dI_dt = a * S * I - b * I
    dR_dt = b * I
    return [dS_dt, dI_dt, dR_dt]
```

Next we want to set our parameters and initial conditions. You can modify these based on the parameters provided in the homework. You won't need to change the tolerances. 

```python
# Time range and initial conditions
a = 0.00005
b = 0.15
t_span = (0, 365)
y0=[4999,1,0] #Initial conditions for Susceptible, Infectious, Recovered stored as a vector y0. 

# Specify error tolerences, that is how accurately our Initial Value Problem will be solved numerically 
rtol = 1e-4
atol = 1e-4

```

THe next cell is where the math occurs. This uses the Runge-Kutta method to numerically solve the system of odes and save the results to a variable called sol. This shouldn't need to change unless you've changed the name of any variables above.

```python
# Solve the system
sol = solve_ivp(sirsys1, t_span, y0, method='RK45', rtol=rtol, atol=atol)
#Note that "Runge-Kutta 45" is a standard numerical method to solve IVPs with specified absolute and relative error tolerances.
```

Now we plot the results. The below code uses matplotlib to plot the 3 components of the numerical solution sol. We will now have more components so you should modify those accordingly. 

```python
# Plotting the results
plt.plot(sol.t, sol.y[0], '.', label='S(t)')
plt.plot(sol.t, sol.y[1], '.', color='r', label='I(t)')
plt.plot(sol.t, sol.y[2], '.', color='g', label='R(t)')
plt.xlabel('Time')
plt.ylabel('Population')
plt.legend()
plt.show()
```
